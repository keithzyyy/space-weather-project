# Work out transform logic

## first inpsect number of unique runs per `(valid_time, location) = assumed to be the identifier for a kindex obs in the real world`


```python
# number of unique runs and analysis times per valid_time and location

duckdb.sql(
    f"""
    SELECT valid_time,
        location,
        COUNT(DISTINCT run_id) AS num_unique_runs,
        COUNT(DISTINCT analysis_time) AS num_unique_analysis_times
    FROM read_parquet('{t1_path}')
    GROUP BY valid_time, location
    HAVING num_unique_runs > 1
    ORDER BY num_unique_runs DESC
    """.strip()
)
```

## can confirm: just take latest runid, analysis_time already unified for each (valid_time, location)

```python
duckdb.sql(
    f"""
    SELECT valid_time,
        location,
        MAX(run_id) AS latest_run_id,
        COUNT(DISTINCT kindex) > 1 AS flag_inconsistent_kindex

        FROM read_parquet('{t1_path}')

        GROUP BY valid_time, location
        HAVING valid_time IS NULL
    """
)
```

```python
duckdb.sql(
    f"""
    SELECT
        valid_time,
        location,
        run_id,
        kindex
        FROM
        read_parquet('{t1_path}') AS kindex_data
        WHERE valid_time IS NULL
"""
)
```

```python
# for each (valid_time, location), obtain kindex corresponding to latest runid

# 1. obtain latest runid first,
# 2. then filter to get kindex values for those latest runids
# 3. then join back to get kindex values for those runids

# corrections (although result will still be the same since NULL = NULL is NOT TRUE in SQL):
# 1. COUNT(DISTINCT kindex) > 1 is too naive.
# -> prefer COUNT(DISTINCT kindex) FILTER (WHERE kindex IS NOT NULL) > 1
# 2. add a WHERE valid_time is NOT NULL in latest_runs CTE to avoid counting NULL valid_time as a separate group

df_transform = duckdb.sql(
    f"""

    WITH
    latest_runs AS (
        SELECT valid_time,
               location,
               MAX(run_id) AS latest_run_id,
               COUNT(DISTINCT kindex) FILTER (WHERE kindex IS NOT NULL) > 1 AS flag_inconsistent_kindex

        FROM read_parquet('{t1_path}')
        WHERE valid_time IS NOT NULL
        GROUP BY valid_time, location
    ),

    kindex_latest AS (
        SELECT
        kindex_data.valid_time,
        kindex_data.location,
        kindex_data.kindex,
        latest_runs.latest_run_id,
        latest_runs.flag_inconsistent_kindex

        FROM

        read_parquet('{t1_path}') AS kindex_data

        INNER JOIN
        latest_runs
            ON kindex_data.valid_time = latest_runs.valid_time 
            AND kindex_data.location = latest_runs.location
            AND kindex_data.run_id = latest_runs.latest_run_id
    )

    SELECT *
    FROM kindex_latest
    ORDER BY flag_inconsistent_kindex DESC;

    """.strip()
    
).to_df()

display(df_transform)
```

```python
# verifying correctness of dedup and flagging logic

# 0. first things first, there should be no duplicate obs for (valid_time, location) in the transformed dataframe,
# since we are only keeping the latest run_id for each (valid_time, location)
display(
    df_transform.\
    groupby(['location', 'valid_time']).\
    size().\
    reset_index(name='count').\
    sort_values('count', ascending=False)
)

# 1. if there are any duplicate kindex values for the same valid_time and location,
# then flag_inconsistent_kindex should be True
display(
    df_transform[df_transform['flag_inconsistent_kindex'] == True])

# 2. check that there are no nulls
display(df_transform.isnull().sum())
```
